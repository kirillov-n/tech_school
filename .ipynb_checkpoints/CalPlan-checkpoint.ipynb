```python
names = df["Наименование и группы"]
uniquenames = names.unique()
names = []
for i in range (len(uniquenames)):
    a = str(uniquenames[i])
    a = re.split('\s', a)[:-1]
    new_name = ' '.join(a)
    names.append(new_name)

unique_names_ = list(set(names))
```

```python
stop_elements = ['', 'Итого по']
unique_names = []
for e in unique_names_:
    if e not in stop_elements:
        unique_names.append(e)
unique_names.sort()
```
